```python
pdf = wand(filename = "Files/10234256.pdf", resolution = 100)
pdfImage = pdf.convert('jpeg')

Blobs = []

for g in pdfImage.sequence:
    page = wand(image = g)
    Blobs.append(page.make_blob('jpeg'))

extract = []

for b in Blobs:
    m = Image.open(io.BytesIO(b))
    text = pytesseract.image_to_string(m, lang = 'eng')
    extract.append(text)

print('Done extracting text')
```

```python
df = pd.DataFrame(extract,columns =['Page_Text'])\
.replace(r'\n',' ', regex=True)
```

```python
df = df.to_string()
```

```python
#columns = ['filename','text']
#data = pd.DataFrame()
#files = pd.DataFrame()
data = []
files = []
```

```python
import glob

for file in glob.glob("DocsOCR/DocsOCR/*.pdf"):
    pdf = wand(filename = file, resolution = 100)
    pdfImage = pdf.convert('jpeg')

    Blobs = []

    for g in pdfImage.sequence:
        page = wand(image = g)
        Blobs.append(page.make_blob('jpeg'))

    extract = []

    for b in Blobs:
        m = Image.open(io.BytesIO(b))
        text = pytesseract.image_to_string(m, lang = 'eng')
        extract.append(text)

    print('Done extracting text')
    
    data.append(str(extract))
    files.append(str(file))
    #data = data.replace(r'\n',' ', regex=True)
```

```python
text = pd.DataFrame(data, columns=['text'])
```

```python
text = text.replace(r"\n",' ', regex=True)
```

```python
text
```

```python
f = pd.DataFrame(files, columns=['filename'])
```

```python
f = f.replace(r"DocsOCR/DocsOCR\\", '', regex=True)
f = f.replace (r".pdf", '', regex=True)
f = f.replace(r"\n",' ', regex=True)

f['filename'] = f['filename'].astype(np.int64)
```

```python
f
```

```python
final = pd.merge (f, text, left_index=True, right_index = True)
```

```python
final = pd.merge (final, df, left_on='filename', right_on='Index')
```
